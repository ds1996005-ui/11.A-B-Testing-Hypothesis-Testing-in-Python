```python
# step 1
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# step 2
df = pd.read_excel(r"C:\Users\HP\Downloads\Task 11.xlsx")
```

```python
print(df.head())
```

```python
# Step 3: Identify groups
control = df[df['group'] == 'Control']
test = df[df['group'] == 'Test']
```

```python
# Step 4: Calculate conversion rates (binary outcome)
control_conv_rate = control['converted'].mean()
test_conv_rate = test['converted'].mean()

print("Control Conversion Rate:", control_conv_rate)
print("Test Conversion Rate:", test_conv_rate)
```

```python
# Step 5: Define hypothesis
# H0: Conversion rates are equal
# H1: Conversion rates are different
alpha = 0.05
```

```python
# Step 6: Choose test
# If binary conversion → two-proportion z-test
# If continuous metric (e.g. revenue) → independent t-test

# Example: two-proportion z-test
n_control = len(control)
n_test = len(test)
success_control = control['converted'].sum()
success_test = test['converted'].sum()

p_pool = (success_control + success_test) / (n_control + n_test)
se = np.sqrt(p_pool * (1 - p_pool) * (1/n_control + 1/n_test))
z_score = (control_conv_rate - test_conv_rate) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
# Step 7: Confidence interval for difference
diff = test_conv_rate - control_conv_rate
ci_low = diff - 1.96 * se
ci_high = diff + 1.96 * se
print("95% CI for difference:", (ci_low, ci_high))
```

```python
# Step 8: Visualization
plt.bar(['Control','Test'], [control_conv_rate, test_conv_rate], color=['blue','green'])
plt.ylabel("Conversion Rate")
plt.title("A/B Test Conversion Rates")
plt.show()
```

```python
# Step 9: Final decision
if p_value < alpha:
    print("Result: Reject H0 → Significant difference.")
else:
    print("Result: Fail to reject H0 → No significant difference.")
```
